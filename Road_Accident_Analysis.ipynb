```python
# How to Select Columns

Road_Accident[['Accident_Severity','Day_of_Week', 'Time']]
```

```python
Road_Accident[
    (Road_Accident['Accident_Severity'] == 'Fatal') &
    (Road_Accident['Day_of_Week'] == 'Saturday') &
    (Road_Accident['Number_of_Casualties'] > 2)
]
```

```python
Road_Accident.query('Accident_Severity == "Fatal" & Day_of_Week == "Saturday" & Number_of_Casualties > 2')
```

```python
Road_Accident.groupby('Accident_Severity')['Accident_ID'].count().reset_index()
```

```python
Road_Accident.groupby(['Accident_Severity', 'Day_of_Week'])['Accident_ID'].count().reset_index()
```
